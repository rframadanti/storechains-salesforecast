```python
#Sales Data are less than 942 rows
missing = train_og[train_og.groupby('Store')['Date'].transform('nunique') < 942]
missing.head()
```

```python
missing.info()
```

```python
#missing.to_csv("train_missing-rows.csv", index=False)
#missing = pd.read_csv('train_missing-rows.csv')
```

```python
missing["Store"].unique()
```

```python
missing["Date"] = pd.to_datetime(missing["Date"].astype(str), format = '%d/%m/%Y')
missing.head()
```

```python
idx = pd.date_range('01/01/2013', '31/07/2015')
df = pd.DataFrame(columns = ['Store','DayOfWeek','Date','Sales','Customers','Open','Promo','StateHoliday','SchoolHoliday'])
for i in missing["Store"].unique():
    # Filter by Store
    subset = missing[(missing["Store"]==i)]
    subset = subset.set_index('Date')
    subset = subset.reindex(idx)
    subset["Store"]=i
    df = pd.concat([df,subset])
```

```python
df = df.drop(["Date"], axis = 1)
df.head()
```

```python
complete = train_og[train_og.groupby('Store')['Date'].transform('nunique') == 942]
complete.info()
```
