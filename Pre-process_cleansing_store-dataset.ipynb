```python
#### CompetitionOpenSinceMonth & Year, fillna with Median, and CompetitionDistance fillna with mean
```

```python
store["CompetitionOpenSinceMonth"]=store["CompetitionOpenSinceMonth"].fillna(store["CompetitionOpenSinceMonth"].median())
```

```python
store["CompetitionOpenSinceMonth"].median(), store["CompetitionOpenSinceYear"].median(), store["CompetitionDistance"].mean()
```

```python
store["CompetitionOpenSinceYear"]=store["CompetitionOpenSinceYear"].fillna(store["CompetitionOpenSinceYear"].median())
```

```python
store["CompetitionDistance"][store["CompetitionDistance"].isna()==True] = store["CompetitionDistance"].fillna(store["CompetitionDistance"]).mean()
```

```python
##Combininng CompetitionOpenSinceYear & Month into one Competition Open Since Date column
## To calculate how many days competitor coexisted & reduce dimension
```

```python
store["CompetitionOpenSinceYear"] = store['CompetitionOpenSinceYear'].astype("int")
store["CompetitionOpenSinceMonth"]=store['CompetitionOpenSinceMonth'].astype("int")
```

```python
store['CompetitionOpenSinceDate'] = pd.to_datetime(store.CompetitionOpenSinceYear.astype(str) + '-' + store.CompetitionOpenSinceMonth.astype(str))
```

```python
########filling NA to Promo2SinceYear and Week  (2016) first week is out of the sales period range so also could mean
######## 

store["Promo2SinceYear"] = store["Promo2SinceYear"].fillna(2016)
store["Promo2SinceWeek"] = store["Promo2SinceWeek"].fillna(1)
```

```python
## 1.0  <-- to remove the decimal point, changing data type to integer.
store["Promo2SinceYear"] = store["Promo2SinceYear"].astype(int)
store["Promo2SinceWeek"] = store["Promo2SinceWeek"].astype(int)
```

```python
## Creating a new column "Promo2SinceDate" by combining Promo2SinceYear & Date
dates = store.Promo2SinceYear*100+store.Promo2SinceWeek

store['Promo2SinceDate'] = pd.to_datetime(dates.astype(str) + '0', format='%Y%W%w')
```

```python
for i in range(len(store.PromoInterval)):
    if store["PromoInterval"][i] == "Jan,Apr,Jul,Oct":
        store["PromoInterval"][i] = 1
        
    elif store["PromoInterval"][i] == "Feb,May,Aug,Nov":
        store["PromoInterval"][i] = 2
        
    elif store["PromoInterval"][i] == "Mar,Jun,Sept,Dec":
        store["PromoInterval"][i] = 3
    
    else: 
        store["PromoInterval"][i] = 0
    
        
```
